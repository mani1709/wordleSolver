# Wordle solver

Run the first code, this setups everything
After that run the second code and follow the instructions

```python
import requests
url = 'https://gist.githubusercontent.com/cfreshman/a03ef2cba789d8cf00c08f767e0fad7b/raw/5d752e5f0702da315298a6bb5a771586d6ff445c/wordle-answers-alphabetical.txt'
page = requests.get(url)
allWords = page.text.split("\n")

letterFrequency = [['a', 0.082], ['b', 0.015], ['c', 0.028], ['d', 0.043], ['e', 0.13], ['f', 0.022], ['g', 0.02], ['h', 0.061], 
                  ['i', 0.07], ['j', 0.0015], ['k', 0.0077], ['l', 0.04], ['m', 0.025], ['n', 0.067], ['o', 0.075], ['p', 0.019], 
                  ['q', 0.00095], ['r', 0.06], ['s', 0.063], ['t', 0.091], ['u', 0.028], ['v', 0.0098], ['w', 0.024],
                  ['x', 0.0015], ['y', 0.02], ['z', 0.0074]]

fiveLetterWords = []

for i in range(len(allWords)):
    currWord = allWords[i].lower()
    if(len(currWord) == 5):
        alreadyUsedLetters = []
        probability = 0;
        for j in range(len(currWord)):
            addedProbability = letterFrequency[ord(currWord[j]) - 97][1]
            if(currWord[j] not in alreadyUsedLetters):
                alreadyUsedLetters.append(currWord[j])
                probability = probability + addedProbability
        fiveLetterWords.append([currWord, probability])
        
def sortingFunction(e):
    return e[1]

def solve(wordList, containingLetters, exactlyPositionedLetters, notContainedLetters):
    containingList = []
    
    for i in range(len(wordList)):
        isValid = True
        for j in range(len(containingLetters)):
            letterContained = containingLetters[j] in wordList[i][0]
            if(letterContained == False):
                isValid = False
                
        if(isValid == True):
            containingList.append(wordList[i])
        
    exactlyPositionedLettersList = []
    
    for i in range(len(containingList)):
        isValid = True
        for j in range(len(exactlyPositionedLetters)):
            if(exactlyPositionedLetters[j] != '.'):
                letterOnCorrectPosition = exactlyPositionedLetters[j] == containingList[i][0][j]
                if(letterOnCorrectPosition == False):
                    isValid = False
                
        if(isValid == True):
            exactlyPositionedLettersList.append(containingList[i])
            
    outList = []
    
    for i in range(len(exactlyPositionedLettersList)):
        isValid = True
        for j in range(len(notContainedLetters)): 
            if(notContainedLetters[j] in exactlyPositionedLettersList[i][0]):
                isValid = False
            
        if(isValid == True):
            outList.append(exactlyPositionedLettersList[i])
            
            
    outList.sort(key = sortingFunction, reverse = True)
        
    return outList
```

```python
# put all existing letters in lettersExisting
# put all not existing letters in notContainingList
# put the known positions in lettersCorrectPositions where . is a whitespace
#   e.g "....e" for a word that has e on the last position

lettersExisting = ""

lettersCorrectPosition = "....."

notContainingList = ""

solve(fiveLetterWords, lettersExisting, lettersCorrectPosition, notContainingList)
```
